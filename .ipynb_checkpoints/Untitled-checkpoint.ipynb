```python
from seleniumwire import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import numpy as np
import cv2
from matplotlib import pyplot as plt
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

import signal
from contextlib import contextmanager
import monkey
import os

class TimeoutException(Exception): pass

@contextmanager
def time_limit(seconds):
    def signal_handler(signum, frame):
        raise TimeoutException("Timed out!")
    signal.signal(signal.SIGALRM, signal_handler)
    signal.alarm(seconds)
    try:
        yield
    finally:
        signal.alarm(0)
```

```python
def get_players():
    user_state = driver.execute_script("return window.user_state")
    player_id = user_state['cloud']['pid']
    game_state = driver.execute_script("return window.game_state")
    players = parse_players(game_state)
    
    try:
        player_self = [p for p in players if str(p["ID"]) == str(player_id)][0]
    except:
        player_self = []
        
    return players, player_self

def parse_players(player):
    
    game_state = driver.execute_script("return window.game_state")
    
    def parse_player(player,pid):
        try:
            return {
                "ID": pid,
                "HP": player["HP"],
                "alive": player["HP"] > 3,
                "class": player["class"],
                "place": player["place"],
                "position_x": player.get("pos", {}).get("x", 0),
                "position_y": player.get("pos", {}).get("y", 0),
                "to_x": player.get("to", {}).get("x", 0),
                "to_y": player.get("to", {}).get("y", 0),
                "inertia_x": player.get("inertia_x", {}).get("x", 0),
                "inertia_y": player.get("inertia_y", {}).get("y", 0),
                "error": False
            }
        except:
            return {
                "ID": pid,
                "error": True
            }

    return [parse_player(game_state["players"][pid],pid) for pid in game_state["players"]]

class Players:
    def __init__(self):
        #P = Blue, R = Pink, S = Yellow
        self.RULES = {'P': 'R', 'R': 'S', 'S': 'P'}
        self.update_players()
    
    def update_players(self):
        self.players, _ = get_players()
        
        if len(_) > 2:
            self.player_self = _
        
    def get_player_by_id(self, t):
        self.update_players()        
        return self.players[t]
    
    def check_color_exist(self):
        self.update_players()
        
    def get_target_idx(self):
        self.update_players()

        for i in range(len(self.players)):
            curr = self.players[i]
            
            if 'class' in curr and 'class' in self.player_self:
                if curr['class'] == self.RULES[self.player_self['class']] and curr['HP'] > 0:
                    return curr
        
        for i in range(len(self.players)):
            curr = self.players[i]
            
            if 'class' in curr and 'class' in self.player_self:
                if curr['class'] == self.player_self['class'] and curr['HP'] > 0:
                    return curr
            
        return {}
```
